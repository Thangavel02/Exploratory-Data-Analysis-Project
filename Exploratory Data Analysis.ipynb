```python
Country_names=df_final.Country.value_counts().index
```

```python
plt.pie(Country_values,labels=Country_names)
```

```python
#Top 3 countries used zomato
plt.pie(Country_values[:3],labels=Country_names[:3])
```

```python
df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating_count'})
```

```python
Rating=df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating_count'})
```

# Observation
Rating from 1.8 - 2.4 = poor
Rating from 2.5 - 3.4 = Average
Rating from 3.5 - 3.9 = Good
Rating from 4.0 - 4.4 = Very good
Rating from 4.5 - 4.9 = Excellent

```python
plt.rcParams['figure.figsize']=(10,10)
sns.barplot(x='Aggregate rating',y='Rating_count',data= Rating)
```

```python
plt.rcParams['figure.figsize']=(10,5)
sns.barplot(x='Aggregate rating',y='Rating_count',data= Rating, hue='Rating color')
```

```python
plt.rcParams['figure.figsize']=(10,5)
sns.barplot(x='Aggregate rating',y='Rating_count',data= Rating, hue='Rating color',palette=['white','red','orange','yellow','green','blue'])
```

```python
#count plot
sns.countplot(x="Rating color",data=Rating, palette=['black','red','orange','yellow','green','blue'])
```
